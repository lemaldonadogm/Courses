# <font color='red' style='font-size: 30px;'>4.2 Avaliando o Modelo Estimado</font>
<hr style='border: 2px solid red;'>

## Avaliando as estatísticas de teste do modelo

```python
print(modelo_stats_models.summary())

# O R2 ajustado leva em conta os graus de liberdade (o número de variáveis explicativas). Como o R2 sempre aumenta se eu aumento
# as variáveis explicativas, o R2 ajustado pode diminuir caso eu aumente o grau de liberdade e essa variável não ajude mais a explicar
# o comportamento da variável dependente. Portanto, ele é melhor de analisar

# Teste F testa se todas os coeficientes das variáveis explicativas são iguais a zero (conjuntamente), ou seja, se faz sentido ou não
# essa regressão. Prob(F-statistic) maiores ou iguais que 0.05 nao rejeita H0, ou seja, a regressão não faze sentido. Nesse caso, 
# a regressão é válida.

# Teste t avalia individualmente cada variavel. H0: coeficiente da variavel é nulo. Para p-valores >= 0.05, não rejeita-se H0, ou seja,
# aquela variável não é estatisticamente significativa. Por exemplo, o Log_Dist_Famarcia tem o p-valor de 0.603, ou seja, não é esta-
# tisticamente significativo para o modelo.
```

### Coeficiente de Determinação - R²

O coeficiente de determinação (R²) é uma medida resumida que diz quanto a linha de regressão ajusta-se aos dados. É um valor entra 0 e 1.

$$R^2(y, \hat{y}) = 1 - \frac {\sum_{i=0}^{n-1}(y_i-\hat{y}_i)^2}{\sum_{i=0}^{n-1}(y_i-\bar{y}_i)^2}$$

## Obtendo o intercepto do modelo

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>O <b>intercepto</b> representa o efeito médio em $Y$ (Preço do Imóveis) tendo todas as variáveis explicativas excluídas do modelo. <p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>Os <b>coeficientes de regressão</b> $\beta_2$ e $\beta_3$ são conhecidos como <b>coeficientes parciais de regressão</b> ou <b>coeficientes parciais angulares</b>. </p>

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>Um aspecto interessante do modelo log-linear, que o tornou muito utilizado nos trabalhos aplicados, é que os coeficientes angulares $\beta_2$ e $\beta_3$ medem as elasticidades de Y em relação a $X_2$ e $X_3$, isto é, a variação percentual de Y correspondente a uma dada variação percentual (pequena) em $X_2$ e $X_3$.</p>
