## Efeito do tamanho das amostras

```python
import matplotlib.pyplot as plt

# Acontece que eu não sei se meu dataframe está ordenado por algum motivo.. então vou ordenar ele aleatoriamente com sample
temp = notas_medias_por_filmes_com_pelo_menos_10_votos.sample(frac=1, random_state=15)

## Outro jeito de utilizar o random_state, é usando o seed do numpy
import numpy as np
np.random.seed(15) #pode colocar qualquer número, igual o random state, que ele vai aleatorizar

medias = list()
for i in range(1, len(temp)):
    medias.append(temp[0:i].mean())
plt.plot(medias)

## Perceba que com poucas amostras (filmes) tenho uma variação muito grande na média..
# A medida que vou aumentando a quantidade de amostras, a média vai se estabilizando.
```

```python
# Outro jeito de fazer exatamente o que eu fiz em cima
medias2 = [temp[0:i].mean() for i in range(1, len(temp))]
plt.plot(medias2)
```

## Qual o intervalo de confiança da média para a média dos filmes que serão votados (da população) ? E não apenas da amostra que eu tenho

```python
from statsmodels.stats.weightstats import zconfint
zconfint(notas_medias_por_filmes_com_pelo_menos_10_votos, alpha = 0.05)
```

Dado esse conjunto de dados, e que é grande suficiente e segue uma distribuição normal, eu tenho um intervalo de confiança de 3,41 e 3,45 para a média dos filmes em geral. 

Diferença do ttest e ztest -> z teste conjunto maior de dados (mais de 30)

```python
from statsmodels.stats.weightstats import DescrStatsW
descr_stats = DescrStatsW(notas_medias_por_filmes_com_pelo_menos_10_votos)
descr_stats.tconfint_mean(alpha = 0.05)
```

Observe que o test t deu muito parecido que o test z... Para amostras pequenas usar o test t

Eu acredito que a notas média do ToyStory seja maior que a nota média de todos os filmes. Será que no mundo real, isso é verdade?

```python
from statsmodels.stats.weightstats import ztest
ztest(notas_toy_story['rating'], value = 3.4320503405352594) # devolve o valor estatístico e o p-value

# Como p-valor é menor que 0.05, eu rejeito a hipótese nula de que a média das notas do Toy Story é igual a média dos filmes do MovieLens com pelo menos 10 votos.
```

### Efeito do tamanho das amostras no teste Z

```python
# Acontece que eu não sei se meu dataframe está ordenado por algum motivo.. então vou ordenar ele aleatoriamente com sample
temp = notas_toy_story['rating'].sample(frac=1, random_state=75421)

def calcula_z_teste(i):
     media = temp[0:i].mean()
     stat, p_value = ztest(temp[0:i], value = 3.4320503405352594)
     return (i, media, p_value)

valores = np.array([calcula_z_teste(i) for i in range (2, len(temp))]) #1 só tem 1 amostra

plt.plot(valores[:,0], valores[:,1]) #eixo x,eixo y #media
plt.plot(valores[:,0], valores[:,2]) #p_valor
plt.hlines(0.05, xmin=0, xmax=len(temp), colors='red') # linha horizontal do p-valor de 0.05

# Observe que com poucas amostras o p_valor varia muito, mas com muitas amostras ele tende a ser constante e portanto
# deve ser usado para amostras grande (n > 50)
```

## Comparação de dois conjuntos amostras

```python
# Intervalo de confiança entre as médias dessas duas distribuições

print(zconfint(notas_toy_story['rating'], notas['rating']))
print(ztest(notas_toy_story['rating'], notas['rating']))

# As notas do toystory é diferente entre 0.28 e 0.56 a mais que a média dos filmes do MovieLens.
# P-Valor de 3.76e-09, ou seja, muito menor que 0.05, então rejeitamos a hipótese nula de que as médias são iguais.
```

```python
# Fazendo ao contrario
print(zconfint(notas['rating'], notas_toy_story['rating']))
print(ztest( notas['rating'], notas_toy_story['rating']))

# Quer dizer que a média dos filmes em geral é menor entre 0.28 e 0.56 do que a média do Toy Story.
```

```python
from scipy.stats import normaltest

_, p = normaltest(notas_toy_story['rating'])
p
# Como p_valor é menor que 0.05, eu rejeito a hipotese de que a nota do toy story segue uma distribuição normal.
# Logo eu não poderia ter usado o test t nem teste z
```

```python
from scipy.stats import ranksums
_,p = ranksums(notas1['rating'], notas593['rating'])
p
# Hipotese nula de que as distribuições são iguais, como p_valor é menor que 0.05, rejeitamos a hipótese nula, ou seja, 
# as distribuições são diferentes. Aqui não falo de média nem desvio padrao.
# Interpretação humana: Em geral, as notas do filme Silêncio dos Inocentes são maiores do que as notas do Toy Story.
```
