```python
import pandas as pd
import seaborn as sns

tmdb = pd.read_csv('tmdb_5000_movies.csv')
```

```python
tmdb.head()
```

```python
tmdb.describe()
```

```python
ax = sns.displot(tmdb['vote_average'], kde = True)
ax.set(title='Distribuição das Frequências das Notas Médias dos Filmes do TMDB')
ax.set_xlabels('Nota Média')
ax.set_ylabels('Frequência')
```

```python
ax = sns.boxplot(tmdb['vote_average'], orient='h')
ax.set_title('Boxplot das Notas Médias dos Filmes do TMDB')
ax.set_xlabel('Nota Média')
```

Estranho ter alguns filmes com nota média de 0... faz sentido?

```python
tmdb.query("vote_average == 0")
```

```python
tmdb.query("vote_average == 10")
```

Os filmes que tem nota média 0 não possuem votos (vote_count). Vamos tirar eles da análise, não faz sentido

```python
tmdb_com_mais_de_10_votos = tmdb.query("vote_count >= 10")
```

```python
tmdb_com_mais_de_10_votos.describe()
```

```python
ax = sns.displot(tmdb_com_mais_de_10_votos['vote_average'], kde = True)
ax.set(title='Distribuição das Frequências das Notas Médias dos Filmes no TMDB com mais de 10 Votos')
ax.set_xlabels('Nota Média')
ax.set_ylabels('Frequência')

## Percebe-se que a distribuição das notas médias dos filmes tem uma cara da distribuição normal, com uma leve assimetria
## à esquerda. 
```

```python
ax = sns.boxplot(tmdb_com_mais_de_10_votos['vote_average'], orient='h')
ax.set_title('Boxplot das Notas Médias dos Filmes no TMDB com mais de 10 Votos')
ax.set_xlabel('Nota Média')
```

Analisaremos também as avaliações do movieslens

```python
notas = pd.read_csv('ratings.csv')
notas.head()
```

```python
notas_por_filme = notas.groupby('movieId').mean()["rating"]
notas_por_filme

## Isso é uma series
```

```python
ax = sns.displot(notas_por_filme.values, kde = True)
ax.set(title='Distribuição das Frequências das Notas Médias dos Filmes no MoviesLens')
ax.set_xlabels('Nota Média')
ax.set_ylabels('Frequência')
```

Muitos receberam nota 5... estranho, vamos analisar

```python
qtd_notas_por_filme = notas.groupby('movieId').count()
qtd_notas_por_filme.query("rating >= 10")
qtd_notas_por_filme_com_mais_de_10_votos = qtd_notas_por_filme.query("rating >= 10")['rating']
qtd_notas_por_filme_com_mais_de_10_votos
```
